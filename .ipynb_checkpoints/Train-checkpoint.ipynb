```python
import json
from os import listdir
from os.path import isfile, join

files = [f for f in listdir('/home/max/Desktop/pySCIL-MetaTaggerTrain/dialogues') if isfile(join('/home/max/Desktop/pySCIL-MetaTaggerTrain/dialogues', f))]
```

```python
turnlens = []
for f in files:
    with open('dialogues/'+f) as handle:
        data = json.load(handle)
        turnlens.append(len(data['turns']))

print(turnlens)
print('Average turns:', sum(turnlens)/(len(turnlens)))
print('Total turns:', sum(turnlens))
```

```python
# Combine all turns into one long list
allTurns = []
for f in files:
    with open('dialogues/'+f) as handle:
        data = json.load(handle)
        for turn in data['turns']:
            allTurns.append(turn)
```

```python
# make document for each meta tag
commMgmtTurns = []
prcsMgmtTurns = []
taskMgmtTurns = []
taskTurns = []

for turn in allTurns:
    if turn['dialog_act'] != '':
        meta = turn['dialog_act'].split(':')[0].lower()
        if meta == 'comm-mgmt':
            commMgmtTurns.append(turn['text'])
        elif meta == 'prcs-mgmt':
            prcsMgmtTurns.append(turn['text'])
        elif meta == 'task-mgmt':
            taskMgmtTurns.append(turn['text'])
        elif meta == 'task':
            taskTurns.append(turn['text'])
print (len(commMgmtTurns),len(prcsMgmtTurns),len(taskMgmtTurns),len(taskTurns))
```

```python
# Create docs
docs = [(turn['text'],turn['dialog_act'].split(':')[0].lower()) for turn in allTurns if turn['dialog_act'] != '']
print(len(docs))
print(852+1265+299+5418)
```

```python
# create train and test set
from sklearn.model_selection import train_test_split
train, test = train_test_split(docs, test_size=0.2, random_state=42)
print(len(train),len(test))
```

```python
# separate into trainX and trainY
trainX, trainY = zip(*train)
testX, testY = zip(*test)
```

```python
# Vectorize each sentence using sklearn
from sklearn.feature_extraction.text import TfidfVectorizer

# setting up counter vectorizer settings
tfidf_vectorizer = TfidfVectorizer(sublinear_tf=True, norm='l2', ngram_range=(1,2))

# applying the vectorizer
tfidf_vectors = tfidf_vectorizer.fit_transform(trainX)
```

```python
import pandas as pd

# get the first vector out (for the first document)
first_vector = tfidf_vectors[0]
# place tf-idf valeus in pandas data frame
df = pd.DataFrame(first_vector.T.todense(), index=tfidf_vectorizer.get_feature_names(), columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
print(df.shape)
print(tfidf_vectors.shape)
```

```python
# test with naive bayes
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(tfidf_vectors, trainY)
```

```python
test_tfidf = tfidf_vectorizer.transform(testX)
```

```python
predicted = clf.predict(test_tfidf)
```
